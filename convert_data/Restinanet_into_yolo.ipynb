```python
import os,cv2
import pandas as pd
import shutil
```

```python
def change(path_save):
    df = pd.read_csv("/home/thinh-do/Workspace/Furniture_detection/data.csv")
    check = []
    for i in range(3001):
        check.append(1)
    for i in range(3000):
        
        if(df.iloc[i,2]!=0 and check[i]==1):
            s= ""
            name_image = df.iloc[i,0].split('/')[-1]
            #path_image = path + "/" + name_image
            pi_save = path_save +"/images/"+name_image
            shutil.copyfile(df.iloc[i,0],pi_save)
            img =cv2.imread(df.iloc[i,0])
            h,w,d =  img.shape
            
            a = round((df.iloc[i,1]+df.iloc[i,3])/(2*w),5)
            b = round((df.iloc[i,2]+df.iloc[i,4])/(2*h),5)
            c = round((-df.iloc[i,1]+df.iloc[i,3])/h,5)
            d = round((-df.iloc[i,2]+df.iloc[i,4])/h,5)
            s = "0,"+str(a)+","+str(b)+","+str(c)+","+str(d)
            name_txt = name_image.replace("jpg","txt")
            path_txt = path_save +"/labels/"+name_txt
            check[i]=0
            #print(s)
            f = open(path_txt,"w+")
            f.write(s)
            for j in range(i+1,3000):
                k=0
                if(df.iloc[i,0]==df.iloc[j,0] and check[j]==1 ):
                    
                    a = round((df.iloc[j,1]+df.iloc[j,3])/(2*w),5)
                    b = round((df.iloc[j,2]+df.iloc[j,4])/(2*h),5)
                    c = round((-df.iloc[j,1]+df.iloc[j,3])/h,5)
                    d = round((-df.iloc[j,2]+df.iloc[j,4])/h,5)
                    k+=1
                    s= "\n"+"0,"+str(a)+","+str(b)+","+str(c)+","+str(d)
                    #if(k>2):
                    #    s= "\n"+"0,"+str(a)+","+str(b)+","+str(c)+","+str(d)
                    check[j]=0
                    print(s)
                    #print("\n")
                    f.write(s)
            f.close()
```

```python

path_save = "/home/viettinh/Desktop/fpt_comp/no mask"
change(path_save)
```
