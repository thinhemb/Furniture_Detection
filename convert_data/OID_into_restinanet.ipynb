```python
import os
import cv2
import pandas as pd
```

```python
label ={}
```

```python
name_file=[]
x1=[]
x2 =[]
x3=[]
x4=[]
labels=[]
def change_name_file(path_lable, path_image):
    i=0
    for file in os.listdir(path_lable):
        #print(file.split('.')[0])
        #name_file.append( "./dataset/images/"+file.replace("txt",'jpg'))
        path = path_lable+"/"+file
        print(path)
        f = open(path,'r',encoding='utf-8')
        data =f.read()
        print(data)
        data_format=data.replace("\n"," ").split(" ")
        #print(len(data_format))
        if(len(data_format)%5==1):
            data_format.remove(data_format[-1])
        #print(data_format)
        #print(len(data_format).type)
        name_image = file.replace("txt",'jpg')

        for i in range(0,int(len(data_format)/5)):
            name_file.append( "/OID/Dataset/validation/"+name_image)
        img =  path_image+"/"+name_image
        print(img)
        # try:
        #     read_img = cv2.imread(img)
        #     (h, w, d) = read_img.shape
            
        # except:
        #     read_img = cv2.imread(img,-1)
        #     (h,w) = read_img.shape
        for i in range(0,len(data_format)):
            if i%5==0:
                if data_format[i]=='Bed':
                    labels.append("Bed")
                elif data_format[i]=='Table':
                    labels.append("Table")
                elif data_format[i]=='Couch':
                    labels.append("Couch")
            elif i%5==1:
                #print(int(h*float(data_format[i])))
                x1.append(data_format[i])
            elif i%5==2:
                x2.append(data_format[i])
            elif i%5==3:
                x3.append(data_format[i])
            elif i%5==4:
                x4.append(data_format[i])
        f.close()
    
    
            
                
```

```python
path='/home/thinh-do/Workspace/Furniture_detection/OID/Dataset/val_lable'
for file in os.listdir(path):
    path_lable = path+"/"+file
    print(file)
    path_image ="/home/thinh-do/Workspace/Furniture_detection/OID/Dataset/validation"
    change_name_file(path_lable,path_image)
label["image"] = name_file
label["x1"] = x1
label["x2"] = x3
label["x3"] = x2
label["x4"] = x4
label["lables"] = labels
    #print(len(x1), len(x2), len(x3), len(x4))
    #print(len(labels))
    #print(len(name_file))
label_train = pd.DataFrame(label)
label_train.head()
label_train.to_csv("/home/thinh-do/Workspace/Furniture_detection/data_val.csv")
label={}
```
